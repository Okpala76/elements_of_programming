```python
## Studtpool Assessment
## Solution to an Element of programming problem 17.2
def brute_force_combinations(score: int, plays: list[int]) -> int:
    counter = 0 ## This holds the count of valid combinations

    def helper(index: int, remaining: int): # a helper funtion to help for itertation
        nonlocal counter

        if index == len(plays): ## The base case 
            if remaining == 0:
                counter += 1
            return

        play_score = plays[index] # we iterate over the list to get individual play scores

        max_times = remaining // play_score # the max time a play score can occur before reaching the limit

        for times in range(max_times + 1): ## we call the helper the amount of times we got the max_times to be
            helper(index + 1, remaining - times * play_score) # recursion

    helper(0, score) # we call the helper fn on the score
    return counter # we return the counter 

num_valid_comb = brute_force_combinations(12, [2,3,7]) # we call the main fn

print("The number of valid combinations is", num_valid_comb) # we print out the counters result with a pretext
```
