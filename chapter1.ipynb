```python
def divUndConquer(arr):
    if len(arr) <= 1:
        return 0
    
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    profit_left = divUndConquer(left)
    profit_right = divUndConquer(right)

    min_left = min(left)
    max_right = max(right)
    cross_profit = max_right - min_left

    return max(profit_left,profit_right, cross_profit, 0)

divUndConquer([715318])


```

```python

def five_hundred_doors():
  doors = 500*[0]

  for i in range(500):
    step = i+1
    for j in range(i,500,step):
      doors[j] = 1 - doors[j]

  result = []
  for i in range(500):
    if doors[i] == 1:
      result.append(i+1)
  return result

five_hundred_doors()
## O(n^2)
```

```python
import math


def five_hundred_doors_optimized():

    result = []

    for i in range(500):
        sqrt_number = math.sqrt(i+1)
        if sqrt_number == int(sqrt_number):
            result.append(i+1)
    return result
five_hundred_doors_optimized()
## O(n^0.5)
```

```python
## offline_sample_optimized
## problem: 6.5
## Optimized: O(n) time, O(1) space

def offline_sample_optimized(arr: list, sample_size: int):
    for current_index in range(sample_size):
        random_index = random.randint(current_index, len(arr)-1)
        arr[current_index], arr[random_index] = arr[random_index], arr[current_index]
    return arr[:sample_size]

offline_sample_optimized(["a","b","c","d","e"], 3)
```

In this optimization we basically puch random selection upfront and then return them this is the whole idea 

```python
## string reversal

def brute_force_string_reversal(str: str):
    reversed_string = ""
    for x in str:
        reversed_string = x + reversed_string

    return reversed_string

brute_force_string_reversal("milk")


## This is O(n^2) becase the longer the string the higer amout of reversed copy call and every call is a n with the alreay eisting reversed sring which is a along as the sting u returned
```

```python
### string reversal with space O(n)-- using lists to cook

def string_reversal_optimized(str: str):
    characters = []
    left = 0
    right = len(str) - 1
    
    for x in str:
        characters.append(x)

    while left < right:
        characters[left], characters[right] = characters[right], characters[left]
        left += 1
        right -= 1
    
    return "".join(characters)

string_reversal_optimized("milk")
```

```python
def reverse_words(sentence: str) -> str:
    characters = list(sentence)

    def reverse_range(left_index: int, right_index: int):
        while left_index < right_index:
            characters[left_index], characters[right_index] = (
                characters[right_index],
                characters[left_index],
            )
            left_index += 1
            right_index -= 1

    reverse_range(0, len(characters) - 1)

    start_index = 0

    for current_index in range(len(characters) + 1):
        if current_index == len(characters) or characters[current_index] == " ":
            reverse_range(start_index, current_index - 1)
            start_index = current_index + 1


    return "".join(characters)


reverse_words("this is you")
```

```python
 ## Brute force implementation

def brute_force_combination_getter(combibation_string: str):

    mapping = {
        "2": "ABC",
        "3": "DEF",
        "4": "GHI",
        "5": "JKL",
        "6": "MNO",
        "7": "PQRS",
        "8": "TUV",
        "9": "WXYZ"
    }

    first_number_combs = mapping[combibation_string[0]]
    second_number_combs = mapping[combibation_string[1]]

    result = []
    for x in first_number_combs:
        for y in second_number_combs:
            result.append(x+y)

    return result

brute_force_combination_getter("22")
```

```python
### exploring iterative  ... same as Bg o as iterative

def phone_combinations_iterative(digits: str):
    if not digits:
        return []

    mapping = {
        "2": "ABC",
        "3": "DEF",
        "4": "GHI",
        "5": "JKL",
        "6": "MNO",
        "7": "PQRS",
        "8": "TUV",
        "9": "WXYZ",
    }

    result = [""]

    for digit in digits:
        next_result = []

        for existing_combination in result:
            for letter in mapping[digit]:
                next_result.append(existing_combination + letter)

        result = next_result

    return result


print(phone_combinations_iterative("2354"))
```
