```python

def brute_force(r):
    max_profit = 0
    n=len(r)

    for i in range(n):
        for j in range(i+1, n):
            profit = r[j]-r[i]
            if profit > max_profit:
                max_profit = profit
    

    print(max_profit)
    return(max_profit)


brute_force([11,2,34,47])

```

```python
def divUndConquer(arr):
    if len(arr) <= 1:
        return 0
    
    mid = len(arr) // 2
    left = arr[:mid]
    right = arr[mid:]

    profit_left = divUndConquer(left)
    profit_right = divUndConquer(right)

    min_left = min(left)
    max_right = max(right)
    cross_profit = max_right - min_left

    return max(profit_left,profit_right, cross_profit, 0)

divUndConquer([715318])


```

```python

def five_hundred_doors():
  doors = 500*[0]

  for i in range(500):
    step = i+1
    for j in range(i,500,step):
      doors[j] = 1 - doors[j]

  result = []
  for i in range(500):
    if doors[i] == 1:
      result.append(i+1)
  return result

five_hundred_doors()
## O(n^2)
```

```python
import math


def five_hundred_doors_optimized():

    result = []

    for i in range(500):
        sqrt_number = math.sqrt(i+1)
        if sqrt_number == int(sqrt_number):
            result.append(i+1)
    return result
five_hundred_doors_optimized()
## O(n^0.5)
```

```python
## offline_sample_brute_force 
## problem: 6.5
## Brute force: O(nk)

import random


def offline_sample_brute_force(arr: list, sample_size: int):
    result = []

    def build_array(start_index: int, current_array: list):
        if len(current_array) == sample_size:
            result.append(current_array.copy())
            return

        for current_index in range(start_index, len(arr)):
            current_array.append(arr[current_index])

            build_array(current_index + 1, current_array)
            current_array.pop()

    build_array(0, [])

    return random.choice(result)

offline_sample_brute_force(["a","b","c","d","e"], 3)
```
